```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('income.csv')

plt.bar(df['Neighborhood'].sort_values().unique(), df.groupby(by=['Neighborhood']).size())
plt.title("Number of Income-Restricted Housing Options per Region")
plt.xlabel("Regions")
plt.ylabel("# of Income-Restructed Housing Options")
plt.xticks(rotation=90)
plt.show()
```
