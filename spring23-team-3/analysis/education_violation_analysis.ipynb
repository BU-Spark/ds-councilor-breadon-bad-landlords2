```python
import pandas as pd

df = pd.read_csv("../data/education/ACSST5Y2020.S1501_Education.csv")
df.describe()
```

```python
# Boston zip codes
# read zipcode as string to keep the leading 0
boston_zipcodes = pd.read_csv("../data/Boston_ZIP_Codes.csv", dtype={"ZIP5": str})
# remove duplicated zipcodes
boston_zipcodes = boston_zipcodes.drop_duplicates("ZIP5")
boston_zipcodes
```

```python
boston_zip = boston_zipcodes["ZIP5"].to_list()
boston_zip
```

```python
# filtered_data
selected_zipcodes = []
for zip in boston_zip:
    selected_zipcodes.append('ZCTA5 ' + zip + '!!Total!!Estimate')
selected_zipcodes

# filtered_data_top20 = filtered_data.filter(items = ['Label (Grouping)'] + selected_zipcodes)
filtered_boston = df[['Label (Grouping)'] + selected_zipcodes]
# filtered_data = df[selected_zipcodes]
filtered_boston.to_csv('../data/education/filtered_boston.csv', index=False)
filtered_boston
```

```python
# education table
df = pd.read_csv("../data/education/education_summary_boston_df.csv")

new_column_values = ['Population_18_24_years', 
                                     'High_shool_graduate_or_higher',
                                     'Population_25_and_over',
                                     'High_shool_graduate_or_higher',
                                     'Population_25_34_years',
                                     'High_shool_graduate_or_higher',
                                     'Population_35_44_years',
                                     'High_shool_graduate_or_higher',
                                     'Population_45_64_years',
                                     'High_shool_graduate_or_higher',
                                     'Population_65_and_over',
                                     'High_shool_graduate_or_higher',
                                     'Total_population_18_and_over',
                                     'High_shool_graduate_or_higher',
                                     'Total_pop_percent_18_and_over',
                                     'High_shool_graduate_or_higher_perc'
                                    ]
df.insert(0, 'Age_by_education_attainment', new_column_values)
df
```

```python
# Transform education table and have a location column same as cities_df
new_df = df.melt(id_vars='Age_by_education_attainment', var_name='Location', value_name='Count')

# Remove the prefixes from the 'Location' column
new_df['Location'] = new_df['Location'].str.replace('ZCTA5 ', '').str.replace('!!Total!!Estimate', '')

new_df
```

```python
# violation table
cities_df = pd.read_csv('../data/zip_df_with_num_buildings.csv', dtype={"Location": str})
cities_df
```

```python
# for abnormal values like 214.29, re-estimate the value based on average
cities_df['violation_percent'] = cities_df['violation_percent'].fillna(cities_df['violation_percent'].mean())
cities_df
```

```python
# Merge cities_df and the transformed education table using Location column
merge_df = pd.merge(new_df, cities_df, left_on='Location', right_on='Location')
merge_df
```

```python
# Visulization
high_school_df = merge_df[merge_df['Age_by_education_attainment'] == 'High_shool_graduate_or_higher_perc']
high_school_df["Count"] = pd.to_numeric(high_school_df["Count"].str.replace("%", ""))

# descending sort by Total violation 
high_school_df = high_school_df.sort_values(by=['violation_percent'], ascending=False)
high_school_df
```

```python
# violation normalized

import matplotlib.pyplot as plt

high_school_df.plot(kind='bar', x='Location', y=['Count', 'violation_percent'], figsize=(10, 6))
plt.title('Correlation between High School Graduates or Higher % vs Violation %')
plt.xlabel('Location')
plt.ylabel('Percentage')
plt.legend(labels=["High School Graduates or Higher %", "Total Violation %"])
plt.show()
```

```python
# Without violation nomalization
import matplotlib.pyplot as plt
high_school_df = high_school_df.sort_values(by=['Total'], ascending=False)

high_school_df.plot(kind='bar', x='Location', y=['Count', 'Total'], figsize=(10, 6))
plt.title('Correlation between High School Graduates or Higher % vs Violation Count')
plt.xlabel('Location')
plt.ylabel('Count')
plt.legend(labels=["High School Graduates or Higher %", "Total Violation Count"])
plt.show()
```
