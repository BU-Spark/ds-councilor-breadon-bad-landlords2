```python
import pandas as pd
import pandas_usaddress

properties_df = pd.read_csv("../data/property_assessments.csv")
properties_df = pandas_usaddress.tag(properties_df, ['SiteAddress'], granularity='single', standardize=True)
properties_df = properties_df[["OWNER1", "SingleLine"]]
properties_df[:20]
```

```python
buildings_df = pd.read_csv("../data/buildings_df.csv")
buildings_df = pandas_usaddress.tag(buildings_df, ['Location'], granularity='single', standardize=True)
buildings_df.head()
```

```python
df = pd.merge(properties_df, buildings_df, on="SingleLine", how="right")
total_df = df.groupby("OWNER1")["Total"].sum()
total_df = total_df.sort_values(ascending=False)
total_df[:20]
```
